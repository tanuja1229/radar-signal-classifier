```python
# Step 1: Generate Synthetic PDW Data
# PDW = Pulse Descriptor Word
# Each pulse has 5 parameters: TOA, RF, DOA, PA, PW

def generate_pdw_data(num_radars=3, pulses_per_radar=50):
    all_pdw = []
    labels = []
    
    for radar_id in range(num_radars):
        # Each radar has fixed RF but varying TOA
        rf = np.random.uniform(500, 1000)      # Radio Frequency (MHz)
        doa = np.random.uniform(0, 120)         # Direction of Arrival
        pa = np.random.uniform(0, 1)            # Pulse Amplitude
        pw = np.random.uniform(0.1, 1)          # Pulse Width
        pri = np.random.uniform(0.01, 0.02)     # Pulse Repetition Interval
        
        for i in range(pulses_per_radar):
            toa = i * pri + np.random.normal(0, 0.0001)  # Time of Arrival with jitter
            pdw = [toa, rf, doa, pa, pw]
            all_pdw.append(pdw)
            labels.append(radar_id)
    
    return np.array(all_pdw), np.array(labels)

# Generate data
pdw_data, labels = generate_pdw_data(num_radars=3, pulses_per_radar=50)

print("PDW Data Shape:", pdw_data.shape)
print("Number of radars:", len(np.unique(labels)))
print("\nFirst 5 PDW entries:")
print(pdw_data[:5])
```
